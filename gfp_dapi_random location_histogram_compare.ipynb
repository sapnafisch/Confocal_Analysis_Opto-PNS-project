```python
#GFP mean fluorescence intensity

#!/usr/bin/env python
# coding: utf-8

# # SC Injected vs WT — Fluorescence Intensity Histogram Analysis
# 
# **Logic:**
# - For each SC animal (C66, C68, P46): 4 sections, each with multiple locations
# - For each section: find the location with the **highest** mean green fluorescence → "right hemisphere"
# - For each section: find the location with the **lowest** mean green fluorescence → "left hemisphere"
# - Animal-level right = mean of 4 section-level right values
# - Animal-level left = mean of 4 section-level left values
# - For WT: pick any location per section, average across 4 sections
# - Green channel identified via embedded LUT metadata
# - Plot histogram: SC right (green), SC left (pink), WT (grey)

# %% Cell 1 — Imports and setup
import os
import numpy as np
import pandas as pd
from tifffile import imread, TiffFile
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# %% Cell 2 — Configuration
BASE_DIR = r"E:\ND2"  # <-- adjust if needed

SC_ANIMALS = {
    'C66': ['C66_SC_102_1', 'C66_SC_102_2', 'C66_SC_102_3', 'C66_SC_102_4'],
    'C68': ['C68_SC_102_1', 'C68_SC_102_2', 'C68_SC_102_3', 'C68_SC_102_4'],
    'P50': ['P50_SC_109_1', 'P50_SC_109_2', 'P50_SC_109_3', 'P50_SC_109_4'],
}

WT_ANIMALS = {
    'WT1': ['WT1_SC_112_1', 'WT1_SC_112_2', 'WT1_SC_112_3', 'WT1_SC_112_4'],
    'WT2': ['WT2_SC_101_1', 'WT2_SC_101_2', 'WT2_SC_101_3', 'WT2_SC_101_4'],
    'WT3': ['WT3_SC_102_1', 'WT3_SC_102_2', 'WT3_SC_102_3', 'WT3_SC_102_4'],
}

# %% Cell 3 — Helper: find TIFF files for a section (recursive search, 40X only)
def find_section_tiffs(base_dir, section_name):
    """
    Recursively search for 40X TIFF files matching a section name.
    Skips any files without '40X' or '40x' in the filename.
    Returns list of (location_id, filepath) tuples.
    """
    results = []
    for root, dirs, files in os.walk(base_dir):
        for f in files:
            if not f.lower().endswith(('.tif', '.tiff')):
                continue
            if section_name not in f:
                continue
            if '40X' not in f and '40x' not in f:
                continue
            filepath = os.path.join(root, f)
            fname_lower = f.lower()
            loc_idx = fname_lower.find('_loc')
            if loc_idx != -1:
                loc_str = fname_lower[loc_idx:loc_idx+6]
            else:
                loc_str = f
            results.append((loc_str, filepath))
    
    if not results:
        print(f"  ⚠ No 40X TIFFs found for: {section_name}")
    else:
        print(f"  {section_name}: {len(results)} locations")
    
    return sorted(results, key=lambda x: x[0])


# %% Cell 4 — LUT-based green channel identification
def lut_color(lut):
    """Classify a single-channel LUT by its dominant color."""
    r, g, b = lut[0].sum(), lut[1].sum(), lut[2].sum()
    if   r == 0 and g > 0  and b > 0:  return 'CYAN'
    elif r > 0  and g == 0 and b == 0:  return 'Red'
    elif r == 0 and g > 0  and b == 0:  return 'Green'
    elif r == 0 and g == 0 and b > 0:   return 'BLUE'
    else:                                return None


def find_green_channel_index(filepath):
    """
    Read TIFF LUT metadata and return the index of the Green channel.
    Returns (green_channel_index, n_channels, {ch_idx: color_name})
    """
    with TiffFile(filepath) as tif:
        metadata = tif.imagej_metadata
        if metadata and 'LUTs' in metadata:
            luts = metadata['LUTs']
            colors = {}
            green_idx = None
            for i, lut in enumerate(luts):
                lut_arr = np.array(lut)
                color = lut_color(lut_arr)
                colors[i] = color
                if color == 'Green':
                    green_idx = i
            return green_idx, len(luts), colors
    return None, None, {}


# %% Cell 5 — Measure mean fluorescence for a single TIFF
def measure_mean_fluorescence(filepath):
    """
    Load TIFF, find green channel via LUT, return mean fluorescence.
    Prints LUT info and shape for verification.
    """
    try:
        green_idx, n_channels, lut_colors = find_green_channel_index(filepath)
        fname = os.path.basename(filepath)
        
        if green_idx is None:
            if lut_colors:
                print(f"    ⚠ No Green LUT in {fname}. LUTs: {lut_colors}. Skipping.")
                return None
            else:
                print(f"    ⚠ No LUT metadata in {fname}. Skipping.")
                return None
        
        img = imread(filepath)
        
        # Extract green channel
        if img.ndim == 2:
            green_2d = img
        elif img.ndim == 3:
            if img.shape[0] <= 5:
                # (C, Y, X)
                green_2d = img[green_idx]
            else:
                # (Z, Y, X) single channel z-stack
                green_2d = np.max(img, axis=0)
        elif img.ndim == 4:
            if img.shape[0] <= 5:
                # (C, Z, Y, X)
                green_2d = np.max(img[green_idx], axis=0)
            elif img.shape[1] <= 5:
                # (Z, C, Y, X)
                green_2d = np.max(img[:, green_idx, :, :], axis=0)
            else:
                print(f"    ⚠ Ambiguous 4D shape {img.shape}, skipping.")
                return None
        elif img.ndim == 5:
            # (T, C, Z, Y, X)
            green_2d = np.max(img[0, green_idx], axis=0)
        else:
            print(f"    ⚠ Unexpected ndim={img.ndim}, shape={img.shape}, skipping.")
            return None
        
        mean_val = np.mean(green_2d.astype(float))
        return mean_val
    
    except Exception as e:
        print(f"    ✗ Error: {e}")
        return None


# %% Cell 6 — Process all animals
def process_all_animals(base_dir, sc_animals, wt_animals):
    all_records = []
    
    print("=" * 70)
    print("PROCESSING SC ANIMALS")
    print("=" * 70)
    
    for animal_id, sections in sc_animals.items():
        print(f"\n  ANIMAL: {animal_id}")
        
        for section_name in sections:
            tiffs = find_section_tiffs(base_dir, section_name)
            
            if not tiffs:
                continue
            
            loc_means = []
            for loc_id, fpath in tiffs:
                mean_val = measure_mean_fluorescence(fpath)
                if mean_val is not None:
                    loc_means.append((loc_id, mean_val, fpath))
            
            if not loc_means:
                print(f"    ⚠ No valid measurements for {section_name}")
                continue
            
            loc_means_sorted = sorted(loc_means, key=lambda x: x[1])
            lowest = loc_means_sorted[0]
            highest = loc_means_sorted[-1]
            
            print(f"    → RIGHT: {highest[0]} = {highest[1]:.2f}  |  LEFT: {lowest[0]} = {lowest[1]:.2f}")
            
            all_records.append({
                'animal': animal_id,
                'group': 'SC',
                'section': section_name,
                'right_hemi_mean': highest[1],
                'right_hemi_loc': highest[0],
                'left_hemi_mean': lowest[1],
                'left_hemi_loc': lowest[0],
                'n_locations': len(loc_means),
            })
    
    print("\n" + "=" * 70)
    print("PROCESSING WT ANIMALS")
    print("=" * 70)
    
    for animal_id, sections in wt_animals.items():
        print(f"\n  ANIMAL: {animal_id}")
        
        for section_name in sections:
            tiffs = find_section_tiffs(base_dir, section_name)
            
            if not tiffs:
                continue
            
            for loc_id, fpath in tiffs:
                mean_val = measure_mean_fluorescence(fpath)
                if mean_val is not None:
                    print(f"    → {loc_id}: {mean_val:.2f}")
                    all_records.append({
                        'animal': animal_id,
                        'group': 'WT',
                        'section': section_name,
                        'right_hemi_mean': mean_val,
                        'right_hemi_loc': loc_id,
                        'left_hemi_mean': mean_val,
                        'left_hemi_loc': loc_id,
                        'n_locations': 1,
                    })
                    break
    
    return pd.DataFrame(all_records)


# %% Cell 7 — Run the analysis
print("Starting analysis...")
print(f"Base directory: {BASE_DIR}\n")

df_sections = process_all_animals(BASE_DIR, SC_ANIMALS, WT_ANIMALS)

print("\n" + "=" * 70)
print("SECTION-LEVEL RESULTS TABLE")
print("=" * 70)
display(df_sections)


# %% Cell 8 — Compute animal-level means
df_animal = df_sections.groupby(['animal', 'group']).agg(
    right_mean=('right_hemi_mean', 'mean'),
    right_std=('right_hemi_mean', 'std'),
    left_mean=('left_hemi_mean', 'mean'),
    left_std=('left_hemi_mean', 'std'),
    n_sections=('section', 'count'),
).reset_index()

print("\n" + "=" * 70)
print("ANIMAL-LEVEL MEANS (averaged across sections)")
print("=" * 70)

for _, row in df_animal.iterrows():
    print(f"\n  {row['animal']} ({row['group']})  —  {int(row['n_sections'])} sections")
    print(f"    Right hemisphere: {row['right_mean']:.2f} ± {row['right_std']:.2f}")
    print(f"    Left hemisphere:  {row['left_mean']:.2f} ± {row['left_std']:.2f}")

display(df_animal)


# %% Cell 9 — Group-level values for plot
sc_right = df_animal[df_animal['group'] == 'SC']['right_mean'].values
sc_left = df_animal[df_animal['group'] == 'SC']['left_mean'].values
wt_vals = df_animal[df_animal['group'] == 'WT']['right_mean'].values

sc_names = df_animal[df_animal['group'] == 'SC']['animal'].values
wt_names = df_animal[df_animal['group'] == 'WT']['animal'].values

print("\n" + "=" * 70)
print("GROUP-LEVEL VALUES FOR PLOT")
print("=" * 70)

print(f"\n  WT (n={len(wt_vals)}):")
for name, v in zip(wt_names, wt_vals):
    print(f"    {name}: {v:.2f}")
print(f"    Mean ± SEM: {np.mean(wt_vals):.2f} ± {np.std(wt_vals)/np.sqrt(len(wt_vals)):.2f}")

print(f"\n  SC Left hemisphere (n={len(sc_left)}):")
for name, v in zip(sc_names, sc_left):
    print(f"    {name}: {v:.2f}")
print(f"    Mean ± SEM: {np.mean(sc_left):.2f} ± {np.std(sc_left)/np.sqrt(len(sc_left)):.2f}")

print(f"\n  SC Right hemisphere (n={len(sc_right)}):")
for name, v in zip(sc_names, sc_right):
    print(f"    {name}: {v:.2f}")
print(f"    Mean ± SEM: {np.mean(sc_right):.2f} ± {np.std(sc_right)/np.sqrt(len(sc_right)):.2f}")


# %% Cell 10 — Plot
fig, ax = plt.subplots(figsize=(5, 5), dpi=300)

positions = [0, 0.7, 1.4]
labels = ['WT', 'SC\n(Left)', 'SC\n(Right)']
colors = ['#888888', '#e377c2', '#2ca02c']
edge_colors = ['#555555', '#c4407a', '#1a7a1a']

means = [np.mean(wt_vals), np.mean(sc_left), np.mean(sc_right)]
sems = [
    np.std(wt_vals) / np.sqrt(len(wt_vals)) if len(wt_vals) > 1 else 0,
    np.std(sc_left) / np.sqrt(len(sc_left)) if len(sc_left) > 1 else 0,
    np.std(sc_right) / np.sqrt(len(sc_right)) if len(sc_right) > 1 else 0,
]

bars = ax.bar(positions, means, yerr=sems, capsize=6,
              color=colors, edgecolor=edge_colors, linewidth=1.2,
              alpha=0.7, width=0.45, error_kw={'linewidth': 1.5})

# Individual data points
np.random.seed(42)
for i, (data, pos) in enumerate(zip([wt_vals, sc_left, sc_right], positions)):
    jitter = np.random.uniform(-0.08, 0.08, size=len(data))
    ax.scatter(pos + jitter, data, color=edge_colors[i], s=50, zorder=5,
              edgecolor='white', linewidth=0.5)

ax.set_ylabel('Mean Green Fluorescence\nIntensity (a.u.)', fontsize=14)
ax.set_xticks(positions)
ax.set_xticklabels(labels, fontsize=13)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='both', labelsize=12)

# Stats: SC right vs WT
if len(sc_right) >= 2 and len(wt_vals) >= 2:
    t_stat, p_val = stats.ttest_ind(sc_right, wt_vals)
    sig = '***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else 'ns'
    y_max = max(max(sc_right), max(wt_vals), max(sc_left))
    bar_y = y_max * 1.1
    ax.plot([positions[0], positions[0], positions[2], positions[2]],
            [bar_y * 0.98, bar_y, bar_y, bar_y * 0.98], color='black', lw=1.2)
    ax.text(positions[1], bar_y * 1.02, f'{sig} (p={p_val:.4f})', ha='center', fontsize=11)
    print(f"\n  Stats: SC Right vs WT — t={t_stat:.3f}, p={p_val:.4f} ({sig})")

plt.tight_layout()
plt.savefig('SC_vs_WT_histogram.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.savefig('SC_vs_WT_histogram.svg', bbox_inches='tight', facecolor='white')
plt.show()

print("\nSaved: SC_vs_WT_histogram.png / .svg")
```
