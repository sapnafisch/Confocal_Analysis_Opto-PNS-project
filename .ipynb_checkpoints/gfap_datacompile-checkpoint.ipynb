```python
import os, re, csv
import pytesseract
from PIL import Image

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

OUTPUT_DIR = r'E:\ND2\4_GFAP_IBA1\tiff_SC\Output'

rows = []
for fname in sorted(os.listdir(OUTPUT_DIR)):
    if not fname.endswith('_quantification.png'):
        continue

    img = Image.open(os.path.join(OUTPUT_DIR, fname))
    w, h = img.size

    # Read title bar
    title_crop = img.crop((0, 0, w, int(h * 0.05)))
    text = pytesseract.image_to_string(title_crop, config='--psm 7')
    print(text.strip())

    # Read GFAP and IBA1 subtracted panel corners for median
    gfap_crop = img.crop((int(w*0.25), int(h*0.55), int(w*0.50), int(h*0.72)))
    iba1_crop = img.crop((int(w*0.75), int(h*0.55), int(w*1.00), int(h*0.72)))
    gfap_text = pytesseract.image_to_string(gfap_crop, config='--psm 6')
    iba1_text = pytesseract.image_to_string(iba1_crop, config='--psm 6')

    gfap_mean   = re.search(r'GFAP[^\d]*([\d.]+)', text)
    iba1_mean   = re.search(r'IBA1[^\d]*([\d.]+)', text)
    gfap_median = re.search(r'median[=\s]*([\d.]+)', gfap_text, re.IGNORECASE)
    iba1_median = re.search(r'median[=\s]*([\d.]+)', iba1_text, re.IGNORECASE)

    rows.append([
        fname.replace('_quantification.png', '.tif'),
        gfap_mean.group(1)   if gfap_mean   else '',
        gfap_median.group(1) if gfap_median else '',
        iba1_mean.group(1)   if iba1_mean   else '',
        iba1_median.group(1) if iba1_median else '',
    ])

csv_path = os.path.join(OUTPUT_DIR, 'GFAP_IBA1_quantification.csv')
with open(csv_path, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Filename', 'GFAP_mean', 'GFAP_median', 'IBA1_mean', 'IBA1_median'])
    writer.writerows(rows)

print(f"Saved: {csv_path}")
```
