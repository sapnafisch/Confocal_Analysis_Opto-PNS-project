```python
import shutil
from pathlib import Path

# ── CONFIGURATION ──────────────────────────────────────────────────────────────

PARENT_DIR = r"E:\ND2\00_NEW"   # Parent folder containing antibody subfolders

DRY_RUN    = False    # Set to False when ready to actually move files
COPY_FILES = False   # Set to True to copy instead of move

# ── MOVE 10X FILES INTO 10X SUBFOLDERS ────────────────────────────────────────

parent = Path(PARENT_DIR)

# Get all immediate subfolders (e.g. CD3_CD4, CD206_TRPV1, etc.)
subfolders = [f for f in parent.iterdir() if f.is_dir()]
print(f"Found {len(subfolders)} subfolders in {PARENT_DIR}\n")

total_moved = 0

for subfolder in sorted(subfolders):
    # Find all 10X files directly inside this subfolder (non-recursive)
    files_10x = [f for f in subfolder.iterdir() if f.is_file() and "10X" in f.name]

    if not files_10x:
        print(f"  {subfolder.name}: no 10X files found, skipping.")
        continue

    dest_dir = subfolder / "10X"
    print(f"  {subfolder.name}: {len(files_10x)} files -> 10X/")

    for f in sorted(files_10x):
        dest = dest_dir / f.name
        if DRY_RUN:
            print(f"    [DRY RUN] {f.name}")
        else:
            dest_dir.mkdir(parents=True, exist_ok=True)
            shutil.copy2(f, dest) if COPY_FILES else shutil.move(str(f), dest)

    total_moved += len(files_10x)

# ── SUMMARY ────────────────────────────────────────────────────────────────────
action = "Would move" if DRY_RUN else ("Copied" if COPY_FILES else "Moved")
print(f"\n{'='*50}")
print(f"  {action} {total_moved} 10X files total across {len(subfolders)} subfolders.")
print(f"{'='*50}")
if DRY_RUN:
    print("\nDry run complete. Set DRY_RUN = False to actually move the files.")
else:
    print("\nDone!")
```
